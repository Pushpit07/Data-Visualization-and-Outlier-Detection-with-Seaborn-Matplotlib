```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2, 20, 1000)
```

```python
m, c = 3, 10
```

```python
y = m*x + c
```

```python
plt.scatter(x, y)
```

```python
noise = np.random.randn(1000)
```

```python
y_mod = y + noise
```

```python
plt.scatter(x, y_mod)
```

```python
noise = np.random.randn(1000) * 5
```

```python
y_mod = y + noise
```

```python
plt.scatter(x, y_mod)
```
